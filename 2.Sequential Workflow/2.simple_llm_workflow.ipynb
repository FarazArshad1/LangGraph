```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model = ChatOpenAI()
```

```python
class LLMState(TypedDict):
    question : str
    answer : str
```

## Nodes

```python
def llm_qa(state : LLMState) -> LLMState:

    # extract the question from state
    question = state['question']

    # form the prompt
    prompt = f'Answer the following question {question}'

    # ask that question to the LLM
    answer = model.invoke(prompt).content

    state['answer'] = answer

    return state
```

```python
# Create Graph
graph = StateGraph(LLMState)

# add nodes
graph.add_node(name = "Simple QA node", node= llm_qa)



# add edges
graph.add_edge(start_key=START, end_key='llm_qa')
graph.add_edge(start_key='llm_qa', end_key=END)


# compile graph
workflow = graph.compile()
```

## Exceute graph

```python
intial_state = {'question' : "How far is moon from earth"}
result = workflow.invoke(input = intial_state)

from pprint import pprint

pprint(result)
```
